## Análise dos sentimentos em relação aos serviços e produtos

```python
#Criado um segundo dataframe para sofrer modificações
df2 = df
df2.head(10)
```

```python
def word_count(titulo, word):
    cont = 0
    titulo = titulo.lower()
    if word in titulo.lower():
        return titulo.count(word)
    else:
        return 0
```

```python
#principais serviços de uma operadora
df2['atendimento'] = df['titulo'].apply(lambda x: word_count(x, 'atendimento'))
df2['cobrança'] = df['titulo'].apply(lambda x: word_count(x, 'cobranca'))
df2['instalação'] = df['titulo'].apply(lambda x:word_count(x, 'instalacao'))
df2['cancelamento'] = df['titulo'].apply(lambda x: word_count(x, 'cancelamento'))
#principais produtos reclamados
df2['plano'] = df['titulo'].apply(lambda x: word_count(x, 'plano'))
df2['serviço'] = df['titulo'].apply(lambda x: word_count(x, 'fixo'))
df2['portabilidade'] = df['titulo'].apply(lambda x: word_count(x, 'internet'))
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
#Função verifica a frequencia das palavras na label título
def contar_palavras(text):  
    token = FreqDist(word_tokenize(text))
    for word, freq in token.items():
        word, freq
    return word, freq
```

```python
#Criada uma lista de palavras ocorridas nos títulos
lista_palavras = []
for frase in df['titulo']:
    palavras = contar_palavras(frase)
    lista_palavras.append(palavras)
```
